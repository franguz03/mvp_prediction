```python
def unique_player(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df['Team'].str.contains(r'^\d+TM$', na=False)]
        row['Team']=df.iloc[-1,:]['Team']
        return row

players = players.groupby(['Player','Year']).apply(unique_player).reset_index(drop=True)
```

### resolving the charlotte problems
